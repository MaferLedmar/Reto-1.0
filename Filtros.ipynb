```python
#INSTALAR LIBRERIAS
import pandas as pd
import io
```

```python
#Crear un Dataframe a partir de fichero Excel
#instala pip install openpyxl
df = pd.read_excel('Inmuebles.xlsx')
df
```

```python
#FILTROS BASICOS PARA VARIABLES CUALITATIVAS
#Filtro por objeto 
filtro1=df[df["Operación"] == "Venta"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
```

```python
#Filtro por filas
#filtro2= df.iloc[300:302,: ]
filtro2=df.iloc[ [80, 1500, 3000] , : ]  #filas no consecutivas 
filtro2
```

```python
filtro2.to_csv("Filtro2.csv")
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , 5:9]
#filtro3=df.iloc[ : , [1,4,6]]  #columnas no consecutivas 
filtro3
```

```python
filtro3.to_csv("Filtro3.csv")
```
